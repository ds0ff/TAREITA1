```python
from datetime import date
from datetime import datetime
from datetime import timedelta
```

```python
Nombre = input('Escriba su nombre: ')
Edad = input ('Escriba su edad: ')
Edad = int(Edad)
fecha = date.today()
nueva_fecha = fecha+timedelta(days=365)
```

```python
texto1= ("{fecha2}"
    '\nQuerida {nombre}'
'\nYa ha pasado un año y ahora tienes {edad} , porfa no abandones la carrera, todo bien :)'
         '\nSé feliz, y cuida A LOS GATOS AAAAAA.').format(nombre=Nombre, edad=Edad+1, fecha2=nueva_fecha)
print(texto1)
```
